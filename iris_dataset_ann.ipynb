```python
import numpy as np
import pandas as pd
%matplotlib inline
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
dir(iris)
```

```python
iris.target
```

```python
# convertinr target data is multi columns into single column
y=iris.target.reshape(-1,1)
y
```

```python
from sklearn.preprocessing import OneHotEncoder
# One Hot encode the class labels
encoder = OneHotEncoder(sparse=False)
encoder
```

```python
y = encoder.fit_transform(y)
y
```

```python
# split the data for trainig and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris.data,y,test_size=0.20)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam,SGD,RMSprop
```

```python
model=Sequential()
model.add(Dense(10,input_shape=(4,),activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(3,activation='softmax',name='output'))
# Adam optimizer with learning rate of 0.001
optimizer = Adam(lr=0.001)
#optimizer=SGD(learning_rate=0.001)
optimizer=RMSprop(learning_rate=0.001)
model.compile(optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
#print('Neural Network Model Summary: ')
print(model.summary())
```

```python
model.fit(x_train, y_train, verbose=2, batch_size=5, epochs=200)
```

```python
model.evaluate(x_test,y_test)
```
